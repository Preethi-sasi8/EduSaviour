```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load dataset

data = pd.read_excel("/content/drive/MyDrive/student_dropout_with_schemes.xlsx")

# Targets

risk_target = data["Risk_Category"]
scheme_target = data["Top_Priority_Scheme"]

# Columns not used as input

drop_cols = [
"Student_ID",
"Risk_Category",
"Dropout_Risk_Label",
"Recommended_Schemes",
"Top_Priority_Scheme",
"Scheme_Application_Status"
]

X = data.drop(columns=drop_cols)

# Convert categorical features

X = pd.get_dummies(X)

# Train-test split

X_train, X_test, y_train_risk, y_test_risk = train_test_split(
X, risk_target, test_size=0.2, random_state=42
)

_, _, y_train_scheme, y_test_scheme = train_test_split(
X, scheme_target, test_size=0.2, random_state=42
)

# Train models

risk_model = RandomForestClassifier(n_estimators=200)
scheme_model = RandomForestClassifier(n_estimators=200)

risk_model.fit(X_train, y_train_risk)
scheme_model.fit(X_train, y_train_scheme)

joblib.dump(risk_model, "risk_model.pkl")
joblib.dump(scheme_model, "scheme_model.pkl")
joblib.dump(X.columns, "model_columns.pkl")

print("Training completed and models saved.")

# Download to local computer
from google.colab import files

files.download("risk_model.pkl")
files.download("scheme_model.pkl")
files.download("model_columns.pkl")
```
